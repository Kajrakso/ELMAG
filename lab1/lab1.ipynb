```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants
```

```python
def plot_mag_field(B_field_function, x_range: np.ndarray, title, a_range: list = []):
    if len(a_range) == 0:
        plt.plot(x_range, B_field_function(x_range))
    else:
        for a in a_range:
            plt.plot(x_range, B_field_function(x_range, a), label=f"a = {a}")
            plt.legend()

    plt.title(title)
    plt.ylabel("B [gauss]")
    plt.xlabel("Avstand fra senter av spolene [m]")
    plt.show()
```

## 2.2.2 Kort spole

Anta at du har en kort spole med N = 330 viklinger og gjennomsnittsradius ⟨R⟩ = 70 mm
med en konstant strøm I = 1,00 A gjennom viklingene.

```python
I: float = 1.00     # A, strøm
N: int = 330        # #, antall spoler
R: float = 70e-3    # m, gjennomsnittsradius

x_min, x_max, step = -20e-2, 20e-2, 0.5e-2                # m, start - stop - steglengde
x_range: np.ndarray = np.arange(x_min, x_max, step)     # m, liste med x-verdier
```

```python
def B_field_short(x):
    """Kort spole. Returnerer i enheten gauss (=10^-4T)"""
    return (scipy.constants.mu_0 * N*I/(2*R) * (1+x**2/R**2)**(-3/2) )*10**4
```
